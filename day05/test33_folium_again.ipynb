## floiom 심화

```python
# date : 20240202
# file : test33_folium_again.ipynb
# desc : folium 지도
import folium
```

```python
## 타일 종류와 마커 표시
## tiles 종류 : openstreermap(기본), cartodb positron, cartodb darK_matter, open topomap
map = folium.Map(location=[35.173999, 129.196649], zoom_start=15,tiles='cartodb positron')
folium.Marker([35.173999, 129.196649], popup='송정해수욕장').add_to(map) # 좌표 마커 추가
map
```

```python
## 마커 종류 변경
map = folium.Map(location=[35.173999, 129.196649], zoom_start=15,tiles='cartodb positron')
folium.Marker([35.173999, 129.196649], popup='송정해수욕장', # 좌표 마커 추가
              icon=folium.Icon(color='red', icon='glyphicon-ok')).add_to(map) # 아이콘 변경
map
```
